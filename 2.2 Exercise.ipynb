```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import dataset
df = pd.read_excel ('world-population.xlsm')
print (df)
```

```python
# Line Chart
ax = plt.gca()

df.plot(kind='line',x='Year',y='Population',ax=ax)
plt.title('World Population by Year')
plt.ylabel('World Population in billions')
plt.show()
```

```python
# Step Chart
ax = plt.gca()

df.plot(kind='line',x='Year',y='Population',drawstyle='steps', ax=ax)
plt.title('World Population by Year')
plt.ylabel('World Population in billions')
plt.show()
```
